```python
numerical_column = 'Marks'
data_Marks = data[numerical_column]
print(data_Marks)


mean_val = data_Marks.mean()
median_val = data_Marks.median()
mode_val = data_Marks.mode()[0]
std_dev = data_Marks.std()
variance = data_Marks.var()
range_val = data_Marks.max()- data_Marks.min()
 
print("\nStatistics for Numerical Column:")
print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
print(f"Standard Deviation: {std_dev}")
print(f"Variance: {variance}")
print(f"Range: {range_val}")
 
plt.figure(figsize=(8, 5))
plt.hist(data_Marks, bins=10, color='skyblue', edgecolor='black')
plt.title(f"Histogram of {numerical_column}")
plt.xlabel(numerical_column)
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x=data_Marks, color='lightgreen')
plt.title(f"Boxplot of {numerical_column}")
plt.show()

q1 = data_Marks.quantile(0.25)
q3 = data_Marks.quantile(0.75)
iqr = q3- q1
lower_bound = q1- 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = data_Marks[(data_Marks < lower_bound) | (data_Marks > upper_bound)]
print("\nOutliers:")
print(outliers)
 
def marks_group(Marks):
  if Marks >85 :
    return 'distinction'
  elif 35<= Marks <= 85:
    return 'first class'
  else:
    return 'fail'

data['Marks Group'] = data['Marks'].apply(marks_group)

categorical_column = 'Marks Group'
data_cat = data[categorical_column]
 
category_counts = data_cat.value_counts()
print("\nCategory Frequencies:")
print(category_counts)
 
plt.figure(figsize=(8, 5))
category_counts.plot(kind='bar', color='coral', edgecolor='black')

plt.title(f"Bar Chart of {categorical_column}")
plt.xlabel(categorical_column)
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8, 5))
category_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90,
colors=sns.color_palette('pastel'))
plt.title(f"Pie Chart of {categorical_column}")
plt.ylabel("") 
plt.show()
```
